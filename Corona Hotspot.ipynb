```python
#Code for red , orange , green hotspot zones in India
rog_map = folium.Map(zoom_starts = 12)
def dis(i,color):
    folium.Marker(
        location = point,
        popup = dataset['District'][i],
        icon = folium.Icon(color = color, icon = 'tint', icon_color = 'white')
    ).add_to(rog_map)

i = 0 
for point in dir:
    if zones[i][0] == 1:
        dis(i , 'orange')
    elif zones[i][1] == 1:
        dis(i , 'darkred')      
    else:
        dis(i , 'green')
    i+=1
```
